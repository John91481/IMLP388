# #單變量分析繪圖（直方圖、條形圖）
> Dataset: NBA_season1718_salary.csv<br>
ref:https://www.kaggle.com/koki25ando/salary

```python
# 讀取數據
data=pd.read_csv('data/NBA_season1718_salary.csv')
data
```

```python
#對讀進來的數據按薪資降序排序並取前10條數據
salary_top10 =(data.sort_values("season17_18", ascending=False)).head(10)# head(10)列10筆，head()列輸前五筆
#data_sort_values():排序
salary_top10
```

```python
#取出球員列數據，作為條形圖橫坐標

player= salary_top10['Player']

#取出球員薪資列，作為條形圖縱坐標

season_salary=salary_top10['season17_18']
```

```python
plt.figure(figsize=(10,6))

plt.xticks(rotation=45)

plt.ylim(season_salary.min()*0.5,season_salary.max()*1.01)

#ci參數表示允許的誤差範圍（控制誤差棒的百分比，在0-100之間）

sns.barplot(x=player,y=season_salary,data=salary_top10,ci=68)
```

## Result

從條形圖中不難看出，勇士隊的當家球星curry資最高，將近達到3500萬美元，位居第二的是勒布朗詹姆斯，也是在3300萬美元以上。上面的條形圖是按球員作為橫坐標，下面我們按球隊匯總，看看哪個球隊的薪資總額最高。

# 練習 - 彙總球隊薪資繪製成直方圖
按球隊匯總，看看哪個球隊的薪資總額最高。

```python
#彙整用gropby
salary_by_team = data.groupby(by='Tm').sum()
salary_by_team
```

```python
#先將資料集用groupby根據球隊分別匯總
salary_by_team=data.groupby(by='Tm').sum().reset_index() #排序後要.reset_index()
salary_by_team=salary_by_team.sort_values(by="season17_18",ascending=False)
tm_top10=salary_by_team[1:10]
tm=tm_top10['Tm']
salary_sum=tm_top10['season17_18']
sns.barplot(x=tm,y=salary_sum,data=tm_top10) #要寫上x=,y=,data=
sns.color_palette('hls',10)
sns.set(style="whitegrid")
```

```python
#
'''
Q: 為什麼需要salary_by_team=data.groupby(by='Tm').sum().reset_index()可以找到Tm欄位；而salary_by_team=data.groupby(by='Tm').sum()卻找不到Tm欄位
A:
在 Pandas 中，groupby() 函式用於將 DataFrame 按照指定的欄位進行分組。然而，groupby() 函式只會將分組欄位作為結果 DataFrame 的索引，而不會將它作為欄位保留在 DataFrame 中。

讓我們來解釋兩種情況：

使用 salary_by_team = data.groupby(by='Tm').sum().reset_index()
在這種情況下，你先對 DataFrame data 進行 groupby() 操作，並使用 sum() 函式來計算每個組的總和。然後，你使用 reset_index() 函式，這將重置索引，並將 'Tm' 欄位從索引中恢復為普通的列。這樣，salary_by_team DataFrame 就包含了 'Tm' 欄位。

使用 salary_by_team = data.groupby(by='Tm').sum()
在這種情況下，你僅僅是對 DataFrame data 進行了 groupby() 和 sum() 操作，但沒有使用 reset_index() 來重置索引。結果是 salary_by_team DataFrame 的索引是 'Tm'，而不是普通的列。因此，在這個 DataFrame 中，'Tm' 實際上成為了索引，而不是保留為普通的欄位。
'''
```

```python
sns.stripplot(x='sex',y='tip',data=tips) #若要使每個項目不同顏色要放入參數 palette=sns.color_palette()
#sns.stripplot(x='sex',y='tip',data=tips,palette=sns.color_palette())
```

```python
sns.stripplot(x='day',y='tip',data=tips,palette=sns.color_palette())
```

```python
sns.stripplot(x='time',y='tip',data=tips,palette=sns.color_palette())
```

```python
sns.stripplot(x='smoker',y='tip',data=tips,palette=sns.color_palette())
```

因為jitter（數據抖動）這個參數的默認值是True,如果我們將其設置為False，效果如圖：


```python
sns.stripplot(x='day',y='tip',data=tips,palette=sns.color_palette(),jitter=False)
```

```python
sns.swarmplot(x='day',y='tip',data=tips,palette=sns.color_palette())
```

在stripplot函數中，我們只看了某一類別對應的數據，而沒有看類中對應其他類的數據分布，舉個例子，比如我們用stripplot看到男性和女性對應的小費數據，但是不知道男性和女性對應的數據中哪些是吸菸的，哪些是不吸菸的，對於這個問題這兩個函數都有一個hue參數，這裡我們就以分簇散點圖函數為例，演示下這個參數的效果，如圖：


```python
sns.swarmplot(x='day',y='tip',hue='sex',data=tips,palette=sns.color_palette()) # hue:多加一個維度的概念
```

上面這個圖就很好的展示了不同性別在不同的星期中付小費數據的分布。

# #繪製盒圖、小提琴圖
seaborn.boxplot()
這個函數主要是繪製出一個箱型圖來反映離群點數據。

```python
sns.boxplot(x='tip',data=tips)
```

```python
sns.boxplot(y='tip',data=tips)
```

```python
sns.boxplot(y='tip',x='day',data=tips)
```

seaborn庫中的很多繪圖方法就是提前給你畫好模板，我們只需要把數據傳進去就得到相應的圖形，
基於此，我門是不是可以在同一個模板上同時繪製分簇圖和箱型圖呢？

```python
sns.boxplot(x="day",y="total_bill",data=tips)
sns.swarmplot(x="day",y="total_bill",data=tips,color=".25")
```

```python
sns.violinplot(y='tip',x='day',data=tips)
```

小提琴圖即反映了數據的離群情況，同時也反映了數據的分布密度。<br>
此函數跟箱型圖一樣，也有hue參數，可以查看第二個分類屬性下的數據分布，如下：


```python
sns.violinplot(y='tip',x='day',hue='sex',data=tips)
```

可以看到，上面的小提琴圖是4對8個，即當設置hue參數為sex時，每個小提琴的數據集是一類數據，那有沒有辦法讓男性對應數據和女性對應數據都作為一個數據集繪製出一個小提琴圖，只是在一個圖中區分男性數據和女性數據呢？答案是肯定的，violinplot()函數提供了一個split參數，可以做到以上需求，如下：


```python
sns.violinplot(y='tip',x='day',hue='sex',data=tips,split=True)
```

除此以外，我們再來看一下inner參數，上面的小提琴圖中內部都是箱型圖，通過inner參數，我們可以設置其他類型，inner可選參數有：box、quartile(四分位)、point、stick。我們依次看下：


```python
sns.violinplot(y='tip',x='day',hue='sex',data=tips,split=True,inner="box")
```

```python
sns.violinplot(y='tip',x='day',hue='sex',data=tips,split=True,inner="quartile")
```

```python
sns.violinplot(y='tip',x='day',hue='sex',data=tips,split=True,inner="point")
```

```python
sns.violinplot(y='tip',x='day',hue='sex',data=tips,split=True,inner="stick")
```

# #結構化展示多維數據
***Visualizing the multidimensional relationships among the samples is as easy as calling sns.pairplot:***

```python
sns.pairplot(tips,hue='sex',height=2.5)
```

使用FacetGrid時，我們會通過一個pandas DataFrame以及控制圖形網格的行、列和顏色的變量名稱來初始化一個對象。這些維度變量（控制行、列和顏色的變量）應該是分類變量或者離散變量，然後這些變量的不同水平組合起來就構成了整個圖形的每一個子圖（facet，在這裡可以理解為我們維度拆解的最小粒度）


這一函數的目標是一步到位地提供一幅完整的成品圖，它在完成繪圖後還會對每個坐標軸添加注釋。

```python
g = sns.FacetGrid(tips,col='time')
```

```python
g.map(plt.hist,"tip")
```

```python
g = sns.FacetGrid(tips,col='sex',hue="smoker")
g.map(plt.scatter,"total_bill","tip",alpha=.7)
g.add_legend()
```

我們可以指定某個seaborn調色板，也可以通過字典將hue變量中的每個分類與其對應的matplotlib顏色傳遞給函數<br>
（這樣就可以隨心所以使用大量的matplotlib支持的色彩）


```python
pal =dict(Lunch="seagreen",Dinner="gray")
g=sns.FacetGrid(tips, hue="time", palette=pal, height=5)
g.map(plt.scatter,"total_bill","tip",s=50,alpha=.7,linewidth=.5,edgecolor='white')
g.add_legend()
```

```python
tips=sns.load_dataset("tips")
g=sns.PairGrid(tips)
g.map(plt.scatter)
```

map_diag()函數是繪製對角線上的單變量子圖，mag_offdiag是繪製對角線以外的兩個變量間的子圖。

```python
tips=sns.load_dataset("iris")
g=sns.PairGrid(tips)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(tips)
g.map_diag(plt.hist,edgecolor="w")
g.map_offdiag(plt.scatter,edgecolor="w",s=40)
```

```python
'''
PairGrid() 需要手動配置每個子圖的內容，這讓你能夠更靈活地控制每個子圖的外觀和內容。你可以使用不同的繪圖函式（如 scatterplot、kdeplot、boxplot 等）來定制每個子圖的類型。

總結：

pairplot() 是一個高階函式，簡單易用，能夠快速繪製多變量圖形，但自定義能力有限。
PairGrid() 是一個較低階的函式，提供了更多的配置選項和自由度，適合用於自定義多變量圖形。
你可以根據需要選擇適合的函式來繪製多變量圖形，如果只是想快速地探索數據的分布和變數之間的關係，使用 pairplot() 是一個不錯的選擇。如果你需要更多自定義的功能，可以使用 PairGrid() 來進行更靈活的設置。
'''
```
